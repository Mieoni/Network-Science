## Find the degree of each node in G1 and draw a histogram: nodes on the x-axis and degree on the y-axis

```python
# This is a sequence generator. Note the square brackets the RHS of assignment statement.
# for n means for all vertices from 1 to n, 
degree_sequence = [d for n, d in G1.degree()] 

G1.degree () 
G1.nodes()
    
degreeHist = {} # create a vector (or think of this as an array)
for d in degree_sequence:
    if d in degreeHist:
        degreeHist[d]+= 1 #if there is a degree d increment the number of occurrances by one
    else:
        degreeHist[d] = 1
print ("degree", "#nodes")
for d in degreeHist:
    print(f"{d:4}  {degreeHist[d]:6}") # the :4 or :6 is number of spaces for that field when printing
    
value_list = (degree_sequence)
plt.hist(value_list, bins=100)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

G1.degree()
```

## Construct a G2 from G1 by deleting edges between nodes whose degree in G1 whose degree is less than 5; For example there is an edge (5,10) in G1,then we need to remove this edge if the degree of 5 and degree of 10 are both less than 5.

```python
G1.add_edges_from(G1.edges())

# Construct G2 from G1
G2 = G1.copy()

# Iterate over the edges in G1 and remove those whose nodes have degree less than 5 in G1
for u, v in G1.edges():
    if G1.degree(u) < 5 and G1.degree(v) < 5:
        G2.remove_edge(u, v)

# Print edges of G2
print(G2.edges())
```

## Construct a subgraph H2 from H1 by removing edges (a,b) such that the degree of either a or b is less than or equal to 5; Note that if the degree of a is 4 and degree of b is 6, the removing (a,b) will make the degree of b 5. This may result in deleting all the edges from coming out of b.

```python
H1.add_edges_from(H1.edges())

# Construct H2 from H1
H2 = H1.copy()

# Iterate over the edges in G1 and remove those whose nodes have degree less than or equal to 5 in H1
for u, v in G1.edges():
    if H1.degree(u) <= 5 or H1.degree(v) <= 5:
        H2.remove_edge(u, v)

# Print edges of H2
print(H2.edges())
```
