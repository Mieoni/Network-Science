# 2. For the following graph below calculate
# a) degree centrality, b) Eigenvector Centrality, c) PageRank Centrality, d) Katz Centrality.

```python
# Required Header
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx
```

```python
UG = nx.Graph()
UG.add_node("a")
UG.add_node("b")
UG.add_node("c")
UG.add_node("d")
UG.add_node("e")
UG.add_node("f")
UG.add_node("g")
UG.add_node("h")
UG.add_node("i")

UG.add_edge("a","b")
UG.add_edge("a","c")
UG.add_edge("a","d")
UG.add_edge("a","e")
UG.add_edge("a","f")
UG.add_edge("b","g")
UG.add_edge("b","i")
UG.add_edge("e","i")
UG.add_edge("f","h")

nx.draw(UG,node_color="black",node_size=200,font_color="white",font_size=10,font_family="Times New Roman",font_weight="bold",with_labels=True)

plt.show()
```

```python
# degree centrality 
print("Determining the degree centrality of the above graph")
print("----------------------------------------------------")
d = nx.degree_centrality(UG)
for v in UG.nodes():
    print(f"{v:2} {d[v]:.3f}")
sort_list = sorted(d.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in UG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(UG, node_color=color_map, with_labels=True)
```

```python
# b) Eigenvector Centrality 
print("Determining the Eigen Vector centrality of the above graph")
print("----------------------------------------------------")

centrality = nx.eigenvector_centrality(UG)
for v, c in centrality.items():
    print(f"{v:2} {c:.3f}")

sort_list = sorted(centrality.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in UG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(UG, node_color=color_map, with_labels=True)
```

```python
# c) PageRank Centrality 
print("Determining the PageRank Centrality of the above graph")
print("----------------------------------------------------")

centrality = nx.pagerank(UG, alpha=0.5, personalization=None, max_iter=100, tol=1e-06, nstart=None, weight='weight', dangling=None)
for v, c in centrality.items():
    print(f"{v:2} {c:.3f}")

sort_list = sorted(centrality.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in UG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(UG, node_color=color_map, with_labels=True)
```

```python
# d) Katz Centrality
print("Determining the Katz centrality of the above graph")
print("----------------------------------------------------")

centrality = nx.katz_centrality_numpy(UG, alpha=0.2, beta=1.0, normalized=True, weight=None)
for v, c in centrality.items():
    print(f"{v:2} {c:.3f}")

sort_list = sorted(centrality.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in UG:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(UG, node_color=color_map, with_labels=True)
```

# 4. Given the following adjacency matrix, compute the adjacency list and edge list. 

```python
UG = nx.Graph()
UG.add_node("1")
UG.add_node("2")
UG.add_node("3")
UG.add_node("4")
UG.add_node("5")
UG.add_node("6")
UG.add_node("7")
UG.add_node("8")
UG.add_node("9")

UG.add_edge("1","2")
UG.add_edge("1","3")
UG.add_edge("2","3")
UG.add_edge("3","4")
UG.add_edge("3","5")
UG.add_edge("4","5")
UG.add_edge("4","6")
UG.add_edge("4","7")
UG.add_edge("5","6")
UG.add_edge("5","7")
UG.add_edge("6","7")
UG.add_edge("6","8")
UG.add_edge("7","8")
UG.add_edge("8","9")

nx.draw(UG,node_color="black",node_size=200,font_color="white",font_size=10,font_family="Times New Roman",font_weight="bold",with_labels=True)

plt.show()
```

```python
# The adjacency list
adj_list = {
    "1":["2","3"],
    "2":["3"],
    "3":["4","5"],
    "4":["5","6","7"],
    "5":["6","7"],
    "6":["6","7"],
    "7":["8"],
    "8":["9"]
}

A = nx.adjacency_matrix(UG)
matA = A.todense()
print(matA)
# Print the adjacency list
print ('Adjacency List')
for line in nx.generate_edgelist(UG, data=False):
    print(line)
```
